```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Importing the dataset
dataset = pd.read_csv('Effects-of-COVID-19-on-trade-1-February31-March-2020-provisional.csv')
```

```python
dataset.Date = pd.to_datetime(dataset.Date)
```

```python
dataset.head(10)
```

```python
#Exploratory Data Analysis
dataset_direction = dataset.groupby('Direction', as_index=False)['Value'].mean().dropna().sort_values('Value', ascending = False)
```

```python
dataset_direction.head(10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
Direction = ['Imports', 'Exports', 'Reimports']
directionseries = pd.DataFrame(dataset[(dataset['Direction'].\
    isin(Direction)][['Date','Direction', 'Value']].\
    dropna().\
    groupby(['Date', 'Direction'])['Direction','Value'].mean().unstack())
directseries.plot(figsize=(15,8), linewidth=3)
plt.show()
```
